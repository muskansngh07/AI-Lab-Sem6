```python
# Import libraries
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score

# Create Play Tennis dataset
data = {
    'Outlook': ['Sunny','Sunny','Overcast','Rain','Rain','Rain','Overcast','Sunny','Sunny','Rain','Sunny','Overcast','Overcast','Rain'],
    'Temperature': ['Hot','Hot','Hot','Mild','Cool','Cool','Mild','Hot','Cool','Mild','Mild','Mild','Hot','Mild'],
    'Humidity': ['High','High','High','High','Normal','Normal','Normal','High','Normal','Normal','Normal','High','Normal','High'],
    'Wind': ['Weak','Strong','Weak','Weak','Weak','Strong','Strong','Weak','Weak','Weak','Strong','Strong','Weak','Strong'],
    'PlayTennis': ['No','No','Yes','Yes','Yes','No','Yes','No','Yes','Yes','Yes','Yes','Yes','No']
}

df = pd.DataFrame(data)

# Convert categorical data to numerical
le = LabelEncoder()

for column in df.columns:
    df[column] = le.fit_transform(df[column])

# Split features and target
X = df.drop('PlayTennis', axis=1)
y = df['PlayTennis']

# Create Decision Tree model
model = DecisionTreeClassifier(criterion='entropy')  # using Information Gain

# Train model
model.fit(X, y)

# Predictions (on same data for simplicity)
y_pred = model.predict(X)

# Accuracy
print("Accuracy:", accuracy_score(y, y_pred))

# Test with new sample
# Example: Sunny, Cool, High, Strong
test = [[le.fit_transform(['Sunny'])[0],
         le.fit_transform(['Cool'])[0],
         le.fit_transform(['High'])[0],
         le.fit_transform(['Strong'])[0]]]

prediction = model.predict(test)

print("Prediction (Play Tennis?):", "Yes" if prediction[0] == 1 else "No")
```
