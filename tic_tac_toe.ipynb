```python
def tic_tac_toe():
    board=[" "]*9
    player="X"

    for turn in range(9):
        print()
        print(f" {board[0]} | {board[1]} | {board[2]} ")
        print("---+---+---")
        print(f" {board[3]} | {board[4]} | {board[5]} ")
        print("---+---+---")
        print(f" {board[6]} | {board[7]} | {board[8]} ")
        print()
        try:
           move=int(input(f"Player {player} enter any number between 1 to 9: "))-1
           if move < 0 or move > 8 or board[move] != " ":
                raise Exception()
           board[move]=player
        except:
            print("Invalid")
            continue
        wins=[(0,1,2),(3,4,5),(6,7,8),(0,3,6),(1,4,7),(2,6,8),(0,4,8),(2,4,6)]
        if any(board[a]==board[b]==board[c]==player for a,b,c in wins):
            print(f"Player {player} wins!")
            return 

        player="O" if player=="X" else "X"

    print("It's a draw!")
           
```

```python
tic_tac_toe()
```
