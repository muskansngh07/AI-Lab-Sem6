```python
def water_jug(c1,c2,target):
    path=[(0,0)]

    j1=j2=0
    while j1!=target and j2!=target:
        if j1==0:
            j1=c1
        elif j2==c2:
            j2=0
        else:
            pour=min(j1,c2-j2)
            j1-=pour
            j2+=pour
        
        path.append((j1,j2))
        if(len(path)>100):
            return None
        

    return path
```

```python
c1=int(input("Enter the capacity of jug 1: "))
c2=int(input("Enter the capacity of jug 2: "))
target=int(input("Enter the target amount: "))
```

```python
path=water_jug(c1,c2,target)
if path:
    print(f"Path to achieve the target amount of {target}L with {c1}L and {c2}L jugs")
    print("Jug 1 \t Jug 2")
    for space in path:
        print(f"{space[0]} \t {space[1]}")
    final_j1,final_j2=path[-1]
    print(f"Goal achieved with {target}L in Jug{1 if final_j1==target else 2 }")
else:
    print("No solution found")
```
